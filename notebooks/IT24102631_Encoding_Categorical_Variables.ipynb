```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler



```

```python

# Load previous
data_path = '/content/drive/My Drive/PGNO-190/preprocessed_after_IT24102586.csv'
df = pd.read_csv(data_path)


# Encode
le = LabelEncoder()
df['category_encoded'] = le.fit_transform(df['category'].fillna('Unknown'))

# EDA: Boxplot of price by category
plt.figure(figsize=(12,6))
sns.boxplot(x='category_encoded', y='discounted_price', data=df)
plt.title('Discounted Price by Category')
plt.xlabel('Encoded Category')
plt.savefig('/content/eda_price_by_category.png')
plt.show()



# Save
save_path = '/content/drive/My Drive/PGNO-190/preprocessed_after_IT24102631.csv'
df.to_csv(save_path, index=False)
```
