```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler



```

```python

# Load previous
data_path = '/content/drive/My Drive/PGNO-190/preprocessed_after_IT24102566.csv'
df = pd.read_csv(data_path)




# Scale
scaler = StandardScaler()
num_cols = ['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']
df[num_cols] = scaler.fit_transform(df[num_cols].fillna(0))  # Fill NaN with 0 for scaling

# EDA: Correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix After Scaling')
plt.savefig('/content/eda_correlation_heatmap.png')
plt.show()

# Save
save_path = '/content/drive/My Drive/PGNO-190/preprocessed_after_IT24102639.csv'
df.to_csv(save_path, index=False)
```
